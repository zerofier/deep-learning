確率変数は$x$は確率密度関数$f_X(x)$に従う時、可逆関数$Y = g(X)$によって変換した確率変数$y$の確率密度関数を$f_Y(y)$を求めると

確率保存:

$$
P(Y \le y)=P(X \le g^{−1}(y))
$$

チェンジ・オブ・バリアブルズの法則によって

$$
f_Y(y) = f_X(x)\left|\frac{dx}{dy}\right|
$$




$f_X(x)$はガウス分布、変換関数が$y = \tanh(x)$の場合

$$
f_Y(y) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(\tanh^{-1}(y) - \mu)^2}{2\sigma^2}\right) \cdot \frac{1}{1-y^2}
$$

```python
import matplotlib.pyplot as plt
import numpy as np


def gaussian(x, mean, stddev):
    return np.exp(-((x - mean) ** 2) / (2. * stddev**2)) / np.sqrt(2. * np.pi * stddev**2)

mean = 0.3
stddev = 0.6

x = np.arange(-3., 3., 0.01)
y = np.tanh(x)

fig, ax = plt.subplots(figsize=(8, 4.5))
ax.grid(True)
ax.plot(x, gaussian(x, mean, stddev), label=f"$\\mathcal{{N}}({mean}, {stddev}^2)$")
ax.plot(y, gaussian(np.arctanh(y), mean, stddev) / (1 - y**2), label="$\\tanh(x)$")
ax.legend()
```

$f_X(x)$はガウス分布、変換関数が$y = 2\tanh(x)$の場合

$$
f_Y(y) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(\tanh^{-1}(\frac{y}{2}) - \mu)^2}{2\sigma^2}\right) \cdot \frac{2}{4-y^2}
$$

```python
import matplotlib.pyplot as plt
import numpy as np


def gaussian(x, mean, stddev):
    return np.exp(-((x - mean) ** 2) / (2.0 * stddev**2)) / np.sqrt(2.0 * np.pi * stddev**2)


mean = 0.3
stddev = 0.6

x = np.arange(-3.0, 3.0, 0.01)
y = 2.0 * np.tanh(x)

fig, ax = plt.subplots(figsize=(8, 4.5))
ax.grid(True)
ax.plot(x, gaussian(x, mean, stddev), label=f"$\\mathcal{{N}}({mean}, {stddev}^2)$")
ax.plot(y, gaussian(np.arctanh(y * 0.5), mean, stddev) * 2.0 / (4.0 - y**2), label="$2\\tanh(x)$")
ax.legend()
```
